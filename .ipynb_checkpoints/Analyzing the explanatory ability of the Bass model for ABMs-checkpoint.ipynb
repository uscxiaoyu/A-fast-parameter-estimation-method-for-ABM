(2) Relationship between p and P

```python
fig = pl.figure(figsize = (16,12))

for i in range(12):
    ax = fig.add_subplot(3,4,i+1)
    X_data = np.array([x[0][0] for x in d_cont[i]])
    Y_data = np.array([x[1][0] for x in d_cont[i]])
    to_fit = pd.DataFrame({'X':X_data,'Y':Y_data})
    results = smf.ols('Y ~ X-1', data=to_fit).fit()
    a = results.params['X']
    r2 = results.rsquared
    
    ax.scatter(X_data,Y_data,c='k',s=10,marker='o',alpha=0.3)   
    ax.plot(np.arange(0,np.max(X_data)+0.005,0.001),np.arange(0,np.max(X_data)+0.005,0.001),'k--',lw = 1)
    textstr = '$R^2 = %.3f$ \n $\mathrm{\hat P} = %.2f\cdot p$'%(r2,a)
    props = dict(facecolor = 'ivory',alpha=0.5)
    ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=12, verticalalignment='top', bbox=props)
    
    ax.set_title('%s'%title_cont[i],fontsize=15,style='italic')
    ax.set_xlim([0,np.max(X_data)+0.001])
    ax.set_ylim([0,np.max(Y_data)+0.001])

    if i == 0 or i == 4 or i == 8:
        ax.set_ylabel(r'$\hat P$',fontsize=15)
    if i >= 8:
        ax.set_xlabel(r'$p$',fontsize=15)
        
    if i in [1,2,3,5,6,7,9,10,11]:
        ax.set_yticklabels([])
    if i < 8:
        ax.set_xticklabels([])
    ax.grid(True)
        
pl.tight_layout()
```

```python
coeff_cont = []
r2_cont = []
for j in range(8):
    p_data = np.array([x[0][0] for x in d_cont[j]])
    q_data = np.array([x[0][1] for x in d_cont[j]])*(3+j)
    
    P_data = np.array([x[1][0] for x in d_cont[j]]) 
    Q_data = np.array([x[1][1] for x in d_cont[j]])    
    M_data = np.array([x[1][2] for x in d_cont[j]])
    
    to_fit = pd.DataFrame({'p':p_data,'q':q_data,'P':P_data,'Q':Q_data,'M':M_data})
    
    results_P = smf.ols('P ~ p+q-1', data=to_fit).fit()
    results_Q = smf.ols('Q ~ p+q-1', data=to_fit).fit()
    results_M = smf.ols('M ~ p+q-1', data=to_fit).fit()
    
    a = [results_P.params['p'],results_P.params['q'],
         results_Q.params['p'],results_Q.params['q'],
         results_M.params['p'],results_M.params['q']]
    
    b = [results_P.rsquared,results_Q.rsquared,results_M.rsquared]
    coeff_cont.append(a)
    r2_cont.append(b)
```
